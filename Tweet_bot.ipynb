```python
def get_tweet_text_full_text_tuple(tweet, logVerbose=True):
    text = ""
    full_text = ""

    if "full_text" in tweet._json:
        if logVerbose:
            print("This tweet has full_text rather than text.", flush=True)
        full_text = tweet._json["full_text"]
        print("full_text: " + full_text.encode('ascii', 'replace').decode(), flush=True)
        if "display_text_range" in tweet._json:
            beginIndex = tweet._json["display_text_range"][0]
            endIndex   = tweet._json["display_text_range"][1]
            if logVerbose:
                print("Found display_text_range: " + "[" + str(beginIndex) + "," + str(endIndex) + "]")
            text = full_text[beginIndex:endIndex]
        else:
            text = full_text
    elif "extended_tweet" in tweet._json and "full_text" in tweet._json["extended_tweet"]:
        if logVerbose:
            print("This tweet has extended_tweet properties.", flush=True)
        full_text = tweet._json["extended_tweet"]["full_text"]
        if logVerbose:
            print("full_text: " + full_text.encode('ascii', 'replace').decode(), flush=True)
        if "display_text_range" in tweet._json["extended_tweet"]:
            beginIndex = tweet._json["extended_tweet"]["display_text_range"][0]
            endIndex   = tweet._json["extended_tweet"]["display_text_range"][1]
            text = full_text[beginIndex:endIndex]
            if logVerbose:
                print("Found display_text_range in extended_tweet: " + "[" + str(beginIndex) + "," + str(endIndex) + "]")
        else:
            text = full_text
    elif "text" in tweet._json:
        if logVerbose:
            print("This is a classic tweet. text is text", flush=True)
        text = tweet._json["text"]
        print("text: " + text, flush=True)
        full_text = text
    else:
        text = tweet.text
        full_text = text
        print("text: " + text, flush=True)

    return (text, full_text)
```

```python
tweets = api.search("frihet")
tweet = tweets[4]
text, full_text = get_tweet_text_full_text_tuple(tweet)
```
